```python
from wordcloud import WordCloud

alt = ''
with open('texts/alt.txt', 'r') as f:
    alt = f.read()
    # remove new lines
    alt = alt.replace('\n', ' ')
```

```python
altCloud = WordCloud(background_color="white", max_words=200, contour_width=3, collocations = False, contour_color='steelblue')
altCloud.generate(alt)
altcloud_svg = altCloud.to_svg(embed_font=True)

f = open('submission-figures/alt-text-wordcloud.svg', 'w')
f.write(altcloud_svg)
f.close()

import matplotlib.pyplot as plt
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16, 9))
plt.imshow(altCloud)
plt.axis("off")
plt.savefig('submission-figures/alt-text-wordcloud.pdf', bbox_inches='tight')
```
