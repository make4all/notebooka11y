```python
runners = ['axe', 'htmlcs']
for record_type in ['error', 'warning', 'notice']:
    required_grid = different_errors[different_errors['Type'] == record_type].pivot(index='Theme', columns=['Runner', 'DetailCode'], values='count')
    
    for i, runner_type in enumerate(runners):
        filtered_grid = required_grid[runner_type]
        print(record_type, runner_type)
        columns = filtered_grid.columns.tolist()
        for updated_column in columns:
            normalizer = filtered_grid[updated_column].max()
            filtered_grid[updated_column] = filtered_grid[updated_column] / normalizer * 100.0
        
        fig, ax = plt.subplots(nrows=1, ncols=1)
        sns.heatmap(filtered_grid, robust=True, linewidths=0.1, linecolor='black', ax=ax, vmin=0, vmax=100)
        ax.set_title(f'[{record_type}] {runner_type}')
        plt.show()
```
